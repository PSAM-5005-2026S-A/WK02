## Conditional Execution

Adding specificity to our processing.

Look for special conditions and situations to execute special code.

For example, let's count the number of wines from California.

We'll set up a counter, then go through all of the wines and increment our counter when we see wines that have a `province` value equal to `"California"`.

In `Python` if we want to check whether two things are the same we use a double-equal symbol `==`.

This is to differentiate from the single-equal which means assignment.

```python
# Initiate a counter
ca_cnt = 0

# Iterate through all wines
for wine in csvdata:
  # Check condition
  if wine[5] == "California":
    # if true, increment counter
    ca_cnt = ca_cnt + 1

# print counter
print(ca_cnt)
```

### Another example

```python
# TODO: Sum of prices of wines from California
```

```python
# TODO: Sum of prices of wines from California
total_price = 0

for wine in csvdata[1:]:
  if wine[5] == "California":
    total_price = total_price + float(wine[4])

print(total_price)
```

### Dynamic Conditional

So far we've been checking for a _static_ condition: "is this wine from California ?", and updating our counter the exact same way for every wine from California (accumulating some value)

We can use a similar set up and pattern of iteration in order to perform something a little more sophisticated and, instead of counting or accumulating, extract a special wine from our list.

Let's say we want to find the most expensive wine on the list.

We break down the task into similar steps as before:
- initiate a variable
- iterate through the wines
- update the variable under some specific condition.

The difference now is that our `if` statement is going to have variables on both sides of the operator, and when the condition is met, we don't just increment a counter, but we remember all of the properties of the wine that triggered the condition.

```python
# TODO: Dynamic conditional: Find most expensive wine
```

```python
# TODO: Dynamic conditional: Find most expensive wine

# assume the first wine is the most expensive
pricey_wine = csvdata[1]

for wine in csvdata[1:]:
  if float(wine[4]) > float(pricey_wine[4]):
    pricey_wine = wine

print(pricey_wine)
```

### Patterns

We're using slightly modified versions of the same pattern to perform different types of operations on our data, like counting certain types of wine or finding specific wines.

Again, this looks like:
- initialize a variable that will eventually hold the answer we are looking for
- go through all of the items on a list / dataset
- check for certain conditions
- update our result variable

We can combine these steps to filter for more specific situations.

### Combinations

Let's find the ***cheapest*** California wine.

We can first create a new list of all wines from California, and then find the one with the smallest price in that list.

```python
# TODO: Dynamic conditional: Find least expensive wine from CA
```

```python
# TODO: Dynamic conditional: Find least expensive wine from CA

# our result will be a list, so we start with an empty list
ca_wines = []

for wine in csvdata:
  if wine[5] == "California":
    ca_wines.append(wine)

print(ca_wines)
```

```python
# assume it's the first one
# using [0] since this list has no header
cheap_ca = ca_wines[0]

for wine in ca_wines:
  if float(wine[4]) < float(cheap_ca[4]):
    cheap_ca = wine

print(cheap_ca)
```

### Combining Conditionals

We can combine the $2$ loops above and check if $2$ or more conditions are true by chaining our `if` conditions with logic operators.

- `and` requires $2$ conditions to be true in order to execute the code inside the `if`
- `or` executes the code if either of the conditions is true

```python
# TODO: with boolean operators
```

```python
# TODO: with boolean operators

# reset to first wine on dataset
cheap_ca = csvdata[1]

for wine in csvdata:
  # check if 2 conditions are true at once
  if (wine[5] == "California") and (float(wine[4]) < float(cheap_ca[4])):
    cheap_ca = wine

print(cheap_ca)
```

## List Functions

Lists have built-in _superpowers_ (functionality) to help us manipulate, transform and process them.

For example, if we have a list with just numbers (remember to use `float()`) we can get the smallest value, largest value and sum of values by using: `min()`, `max()` and `sum()` respectively.

We call them like this:
`max(list_of_numbers)`

```python
# TODO: extract all prices from CA wines
# TODO: report number of wines, and min, max, sum of prices
```

```python
# TODO: extract all prices from CA wines

# our result is a list of numbers, so start an empty list
ca_prices = []

for row in csvdata[1:]:
  # append price of CA wines
  if row[5] == "California":
    ca_prices.append(float(row[4]))

print(min(ca_prices), ",", max(ca_prices))
print(sum(ca_prices), ",", len(ca_prices))
```
